```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "data/preprocessed_data.csv"
df = pd.read_csv(file_path)

# Check the distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(x=df["Bankrupt?"], palette="viridis")
plt.title("Distribution of Target Variable (Bankrupt?)")
plt.xlabel("Bankrupt (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

# Display summary statistics without using ace_tools
df_summary = df.describe()

# Show the summary statistics
df_summary
```

## Exploratory Data Analysis (EDA) Summary
- Target Variable Distribution:

- The "Bankrupt?" variable is somewhat imbalanced with around 33% bankrupt (1) and 67% non-bankrupt (0).
This may require techniques like balanced sampling or weighted loss functions in machine learning models.
Correlation Analysis:

- The heatmap reveals that some features have moderate correlations.
There may be redundant or highly correlated features that could be removed or transformed to reduce dimensionality.
Feature Distributions:

- Many features have extreme values, suggesting the presence of outliers.
Standardization appears to have been applied, as the means are centered around zero.
Next Steps:

- Identify the most important features for prediction (feature selection).
Handle outliers (e.g., winsorization or robust scaling).
Address the class imbalance if needed.
Train initial models to evaluate predictive performance.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

# Split the data into features (X) and target (y)
X = df.drop(columns=["Bankrupt?"])
y = df["Bankrupt?"]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# Train a RandomForest classifier to compute feature importances
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Get feature importances
feature_importances = rf.feature_importances_

# Create a DataFrame for feature importances
feature_importance_df = pd.DataFrame({
    "Feature": df.columns[:-1],  # Exclude target variable
    "Importance": feature_importances
}).sort_values(by="Importance", ascending=False)

# Display feature importance rankings without ace_tools
importances = feature_importance_df.head(10)  # Show top 10 features

# Plot feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x=importances["Importance"], y=importances["Feature"], palette="viridis")
plt.xlabel("Feature Importance Score")
plt.ylabel("Feature Index")
plt.title("Top 10 Most Important Features")
plt.show()

# Show the dataframe with all feature importances
feature_importance_df.head(20)

```

```python
# Load the original unprocessed dataset
original_file_path = "data/data.csv"
df_original = pd.read_csv(original_file_path)

# Display the first few rows to check feature names
df_original.head()


# Extract feature names from the original dataset (excluding target variable "Bankrupt?")
original_feature_names = df_original.columns.tolist()
original_feature_names.remove("Bankrupt?")

# Extract feature indices from the preprocessed dataset (excluding "Bankrupt?")
preprocessed_feature_indices = df.columns[:-1].astype(int)  # Convert numerical column names to integers

# Map preprocessed feature indices to original feature names
feature_mapping = {idx: original_feature_names[idx] for idx in preprocessed_feature_indices}

# Convert mapping to a DataFrame for better visualization
feature_mapping_df = pd.DataFrame(list(feature_mapping.items()), columns=["Preprocessed Feature Index", "Original Feature Name"])

# Display the mapping
feature_mapping_df


# Ensure that the mapping keys are integers
feature_mapping = {int(k): v for k, v in feature_mapping.items()}

# Update feature importance dataframe with correct feature names
feature_importance_df["Original Feature Name"] = feature_importance_df["Feature"].astype(int).map(feature_mapping)

# Select top 10 most important features with correct names
top_features = feature_importance_df.head(10)

# Plot updated feature importance graph
plt.figure(figsize=(12, 6))
sns.barplot(x=top_features["Importance"], y=top_features["Original Feature Name"], palette="viridis")
plt.xlabel("Feature Importance Score")
plt.ylabel("Feature Name")
plt.title("Top 10 Most Important Features for Bankruptcy Prediction")
plt.show()

# Display the updated feature importance dataframe
feature_importance_df.head(20)
```
