```python
# 2. Boxplots for numeric columns (in batches to avoid overcrowding)
if len(numeric_cols) > 0:
    max_plots_per_fig = 6  # maximum number of boxplots per figure
    for i in range(0, len(numeric_cols), max_plots_per_fig):
        subset = numeric_cols[i:i+max_plots_per_fig]
        num_plots = len(subset)
        fig, axes = plt.subplots(1, num_plots, figsize=(5*num_plots, 6))
        # If only one subplot, put it in a list for iteration
        if num_plots == 1:
            axes = [axes]
        for ax, col in zip(axes, subset):
            ax.boxplot(df[col].dropna())
            ax.set_title(col)
        fig.suptitle(f"Boxplots for Numeric Columns (Batch {i//max_plots_per_fig + 1})")
        plt.tight_layout()
        plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
if len(numeric_cols) > 0:
    max_plots_per_fig = 12  # Define number of histograms per figure
    for i in range(0, len(numeric_cols), max_plots_per_fig):
        subset = numeric_cols[i:i+max_plots_per_fig]
        n_plots = len(subset)
        cols_per_row = 4  # e.g., 4 columns per row
        rows = math.ceil(n_plots / cols_per_row)
        
        fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row * 5, rows * 4))
        # Flatten axes array for easy iteration
        if n_plots == 1:
            axes = [axes]
        else:
            axes = axes.flatten()
            
        for j, col in enumerate(subset):
            axes[j].hist(df[col].dropna(), bins=30)
            axes[j].set_title(col)
        # Remove any unused subplots if they exist
        for k in range(n_plots, len(axes)):
            fig.delaxes(axes[k])
        
        fig.suptitle(f"Histograms for Numeric Columns (Batch {i//max_plots_per_fig + 1})", fontsize=16)
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
target_col = "Bankrupt?"
X = df.drop(columns=[target_col])
y = df[target_col]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a Random Forest Classifier with 100 trees
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the classifier
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

# Perform t-SNE (2 components)
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Create subplots for PCA and t-SNE projections
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# PCA scatter plot
sc1 = axes[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.7)
axes[0].set_title("PCA Projection")
axes[0].set_xlabel("PC1")
axes[0].set_ylabel("PC2")
plt.colorbar(sc1, ax=axes[0], label=target_col)

# t-SNE scatter plot
sc2 = axes[1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', alpha=0.7)
axes[1].set_title("t-SNE Projection")
axes[1].set_xlabel("Component 1")
axes[1].set_ylabel("Component 2")
plt.colorbar(sc2, ax=axes[1], label=target_col)

plt.tight_layout()
plt.show()
```
